```python
# ============================
# IMPORT LIBRARY
# ============================
import simpy
import random
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import pandas as pd
from dataclasses import dataclass
from typing import List, Tuple
import statistics

print("‚úÖ Library berhasil diimport")
```

```python
# ============================
# KONFIGURASI SIMULASI
# ============================

from dataclasses import dataclass
import random
import numpy as np

@dataclass
class Config:
    """Konfigurasi parameter simulasi sistem piket ompreng"""

    # =====================
    # PARAMETER DASAR
    # =====================
    NUM_MEJA: int = 60
    MAHASISWA_PER_MEJA: int = 3

    # =====================
    # PETUGAS
    # =====================
    PETUGAS_LAUK: int = 3
    PETUGAS_ANGKUT: int = 2
    PETUGAS_NASI: int = 2

    # =====================
    # WAKTU PROSES (DETIK)
    # =====================
    LAUK_MIN_TIME: float = 30
    LAUK_MAX_TIME: float = 60

    ANGKUT_MIN_TIME: float = 20
    ANGKUT_MAX_TIME: float = 60

    ANGKUT_MIN_LOAD: int = 4
    ANGKUT_MAX_LOAD: int = 7

    NASI_MIN_TIME: float = 30
    NASI_MAX_TIME: float = 60

    # =====================
    # WAKTU MULAI
    # =====================
    START_HOUR: int = 7
    START_MINUTE: int = 0

    # =====================
    # RANDOM SEED
    # =====================
    RANDOM_SEED: int = 42

    # =====================
    # TOTAL OMPRENG
    # =====================
    @property
    def TOTAL_OMPRENG(self):
        return self.NUM_MEJA * self.MAHASISWA_PER_MEJA


# ============================
# INISIALISASI CONFIG
# ============================

config = Config()

random.seed(config.RANDOM_SEED)
np.random.seed(config.RANDOM_SEED)

print("‚úÖ Konfigurasi simulasi siap")
```

```python
# ============================
# MODEL SIMULASI (FIX FINAL)
# ============================

class PiketOmprengDES:

    def __init__(self, config: Config):
        self.config = config
        self.env = simpy.Environment()

        # RESOURCE
        self.petugas_lauk = simpy.Resource(self.env, capacity=config.PETUGAS_LAUK)
        self.petugas_angkut = simpy.Resource(self.env, capacity=config.PETUGAS_ANGKUT)
        self.petugas_nasi = simpy.Resource(self.env, capacity=config.PETUGAS_NASI)

        # STORE
        self.antrian_lauk = simpy.Store(self.env)
        self.antrian_nasi = simpy.Store(self.env)

        # STATISTIK
        self.statistics = {'ompreng_data': []}
        self.completed = 0

        # UTILISASI
        self.busy_time_lauk = 0
        self.busy_time_angkut = 0
        self.busy_time_nasi = 0

        self.start_time = datetime(
            2024, 1, 1,
            config.START_HOUR,
            config.START_MINUTE
        )

        random.seed(config.RANDOM_SEED)
        np.random.seed(config.RANDOM_SEED)

    # =====================
    # UTIL
    # =====================
    def waktu_ke_jam(self, detik):
        return self.start_time + timedelta(seconds=detik)

    def waktu_lauk(self):
        return random.uniform(self.config.LAUK_MIN_TIME,
                              self.config.LAUK_MAX_TIME)

    def waktu_angkut(self):
        return random.uniform(self.config.ANGKUT_MIN_TIME,
                              self.config.ANGKUT_MAX_TIME)

    def waktu_nasi(self):
        return random.uniform(self.config.NASI_MIN_TIME,
                              self.config.NASI_MAX_TIME)

    # =====================
    # PROSES LAUK
    # =====================
    def proses_ompreng(self, ompreng_id):

        waktu_datang = self.env.now

        request_time = self.env.now
        with self.petugas_lauk.request() as req:
            yield req

            mulai_lauk = self.env.now
            waiting_lauk = mulai_lauk - request_time

            service_time = self.waktu_lauk()
            yield self.env.timeout(service_time)

        selesai_lauk = self.env.now
        self.busy_time_lauk += service_time

        yield self.antrian_lauk.put({
            'id': ompreng_id,
            'datang': waktu_datang,
            'mulai_lauk': mulai_lauk,
            'selesai_lauk': selesai_lauk,
            'waiting_lauk': waiting_lauk
        })

    # =====================
    # PROSES ANGKUT
    # =====================
    def proses_angkut(self):

        while True:

            # STOP condition
            if self.completed >= self.config.TOTAL_OMPRENG:
                break

            if len(self.antrian_lauk.items) == 0:
                yield self.env.timeout(0.5)
                continue

            batch = []
            kapasitas = random.randint(
                self.config.ANGKUT_MIN_LOAD,
                self.config.ANGKUT_MAX_LOAD)

            for _ in range(min(kapasitas, len(self.antrian_lauk.items))):
                batch.append((yield self.antrian_lauk.get()))

            with self.petugas_angkut.request() as req:
                yield req
                waktu = self.waktu_angkut()
                yield self.env.timeout(waktu)

            self.busy_time_angkut += waktu
            selesai_angkut = self.env.now

            for item in batch:
                item['selesai_angkut'] = selesai_angkut
                yield self.antrian_nasi.put(item)

    # =====================
    # PROSES NASI
    # =====================
    def proses_nasi(self):

        while True:

            # STOP condition
            if self.completed >= self.config.TOTAL_OMPRENG:
                break

            if len(self.antrian_nasi.items) == 0:
                yield self.env.timeout(0.5)
                continue

            item = yield self.antrian_nasi.get()

            request_time = self.env.now
            with self.petugas_nasi.request() as req:
                yield req

                mulai_nasi = self.env.now
                waiting_nasi = mulai_nasi - request_time

                service_time = self.waktu_nasi()
                yield self.env.timeout(service_time)

            selesai_nasi = self.env.now
            self.busy_time_nasi += service_time

            self.statistics['ompreng_data'].append({
                'id': item['id'],
                'datang': item['datang'],
                'mulai_lauk': item['mulai_lauk'],
                'selesai_lauk': item['selesai_lauk'],
                'selesai_angkut': item['selesai_angkut'],
                'mulai_nasi': mulai_nasi,
                'selesai_nasi': selesai_nasi,
                'waiting_lauk': item['waiting_lauk'],
                'waiting_nasi': waiting_nasi,
                'total_waktu': selesai_nasi - item['datang'],
                'jam_selesai': self.waktu_ke_jam(selesai_nasi)
            })

            self.completed += 1

    # =====================
    # GENERATOR
    # =====================
    def generate_ompreng(self):
        for i in range(self.config.TOTAL_OMPRENG):
            self.env.process(self.proses_ompreng(i))
            yield self.env.timeout(0)

    # =====================
    # RUN
    # =====================
    def run_simulation(self):

        self.env.process(self.generate_ompreng())
        self.env.process(self.proses_angkut())
        self.env.process(self.proses_nasi())

        self.env.run()
        return self.analyze_results()

      # =====================
    # ANALISIS
    # =====================
    def analyze_results(self):
    
        df = pd.DataFrame(self.statistics['ompreng_data'])
    
        # waktu selesai terakhir (detik simulasi)
        total_time = df['selesai_nasi'].max()
    
        # =====================
        # DISTRIBUSI PER JAM
        # =====================
        distribusi_jam = (
            df['jam_selesai']
            .dt.hour
            .value_counts()
            .sort_index()
            .to_dict()
        )
    
        # =====================
        # HASIL UTAMA
        # =====================
        results = {
            'total_ompreng': len(df),
            'waktu_selesai_detik': total_time,
            'jam_selesai': self.waktu_ke_jam(total_time),
    
            # statistik waktu
            'avg_total_time': df['total_waktu'].mean(),
            'max_total_time': df['total_waktu'].max(),
            'min_total_time': df['total_waktu'].min(),
    
            'avg_wait_lauk': df['waiting_lauk'].mean(),
            'avg_wait_nasi': df['waiting_nasi'].mean(),
    
            # distribusi
            'distribusi_jam': distribusi_jam
        }
    
        # =====================
        # UTILISASI PETUGAS
        # =====================
        results['utilisasi'] = {
            'lauk': (self.busy_time_lauk /
                     (total_time * self.config.PETUGAS_LAUK)) * 100,
    
            'angkut': (self.busy_time_angkut /
                       (total_time * self.config.PETUGAS_ANGKUT)) * 100,
    
            'nasi': (self.busy_time_nasi /
                     (total_time * self.config.PETUGAS_NASI)) * 100
        }
    
        return results, df

print("‚úÖ Model simulasi siap (final)")
```

```python
# ============================
# VISUALISASI DAN ANALISIS
# ============================

def run_scenario(num_meja: int,
                 petugas_lauk: int,
                 petugas_angkut: int,
                 petugas_nasi: int):

    print(f"\n{'='*60}")
    print("SIMULASI SISTEM PIKET OMPRENG IT DEL")
    print(f"{'='*60}")
    print(f"Jumlah Meja        : {num_meja}")
    print(f"Total Ompreng      : {num_meja * 3}")
    print(f"Petugas Lauk       : {petugas_lauk}")
    print(f"Petugas Angkut     : {petugas_angkut}")
    print(f"Petugas Nasi       : {petugas_nasi}")
    print(f"{'='*60}")

    # =====================
    # KONFIGURASI
    # =====================
    config = Config(
        NUM_MEJA=num_meja,
        PETUGAS_LAUK=petugas_lauk,
        PETUGAS_ANGKUT=petugas_angkut,
        PETUGAS_NASI=petugas_nasi
    )

    # =====================
    # RUN MODEL
    # =====================
    model = PiketOmprengDES(config)
    results, df = model.run_simulation()

    if results:

        print("\nüìä HASIL SIMULASI:")
        print(f"Jam mulai: {model.start_time.strftime('%H:%M:%S')}")
        print(f"Jam selesai terakhir: {results['jam_selesai'].strftime('%H:%M:%S')}")
        print(f"Durasi total: {results['waktu_selesai_detik']/60:.2f} menit")

        print("\n‚è±Ô∏è STATISTIK WAKTU:")
        print(f"  Rata-rata total: {results['avg_total_time']:.2f} detik")
        print(f"  Maksimum: {results['max_total_time']:.2f} detik")
        print(f"  Minimum : {results['min_total_time']:.2f} detik")

        print("\nüìà UTILISASI PETUGAS:")
        for bagian, util in results['utilisasi'].items():
            print(f"  {bagian.capitalize():10s}: {util:.1f}%")

        print("\nüïê DISTRIBUSI PER JAM:")
        for jam in sorted(results['distribusi_jam'].keys()):
            print(f"  Jam {jam:02d}:00 - {results['distribusi_jam'][jam]:3d} ompreng")

    return results, df, model

# ============================
# ANALISIS SENSITIVITAS
# ============================

def sensitivity_analysis():

    print("\n" + "="*80)
    print(f"{'ANALISIS SENSITIVITAS SISTEM PIKET OMPRENG':^80}")
    print("="*80)

    scenarios = [

        # ===== SKENARIO DASAR (SOAL) =====
        {"meja":60, "lauk":3, "angkut":2, "nasi":2},

        # ===== VARIASI PETUGAS =====
        {"meja":60, "lauk":2, "angkut":2, "nasi":2},  # kurang petugas lauk
        {"meja":60, "lauk":4, "angkut":2, "nasi":2},  # tambah petugas lauk

        {"meja":60, "lauk":3, "angkut":1, "nasi":2},  # bottleneck angkut
        {"meja":60, "lauk":3, "angkut":3, "nasi":2},  # angkut diperbanyak

        {"meja":60, "lauk":3, "angkut":2, "nasi":1},  # bottleneck nasi
        {"meja":60, "lauk":3, "angkut":2, "nasi":3},  # nasi diperbanyak

        # ===== VARIASI BEBAN =====
        {"meja":40, "lauk":3, "angkut":2, "nasi":2},  # lebih sedikit meja
        {"meja":80, "lauk":3, "angkut":2, "nasi":2},  # lebih banyak meja

        # ===== STRESS TEST =====
        {"meja":80, "lauk":2, "angkut":1, "nasi":1},
        {"meja":80, "lauk":4, "angkut":3, "nasi":3},
    ]

    summary = []

    for i, sc in enumerate(scenarios):

        config = Config(
            NUM_MEJA=sc["meja"],
            PETUGAS_LAUK=sc["lauk"],
            PETUGAS_ANGKUT=sc["angkut"],
            PETUGAS_NASI=sc["nasi"]
        )

        model = PiketOmprengDES(config)
        results, df = model.run_simulation()

        summary.append({
            "Scenario": i+1,
            "Meja": sc["meja"],
            "Total Ompreng": sc["meja"] * config.MAHASISWA_PER_MEJA,
            "Petugas Lauk": sc["lauk"],
            "Petugas Angkut": sc["angkut"],
            "Petugas Nasi": sc["nasi"],
            "Jam Selesai": results["jam_selesai"].strftime("%H:%M:%S"),
            "Avg Total (detik)": round(results["avg_total_time"],2),
            "Util Lauk (%)": round(results["utilisasi"]["lauk"],1),
            "Util Angkut (%)": round(results["utilisasi"]["angkut"],1),
            "Util Nasi (%)": round(results["utilisasi"]["nasi"],1),
        })

    summary_df = pd.DataFrame(summary)

     # ============================
    # FORMAT TAMPILAN
    # ============================
    
    summary_df = summary_df.round({
        "Avg Total (detik)": 2,
        "Util Lauk (%)": 1,
        "Util Angkut (%)": 1,
        "Util Nasi (%)": 1
    })
    
    # pastikan urutan tetap berdasarkan nomor skenario
    summary_df = summary_df.sort_values(
        by="Scenario"
    ).reset_index(drop=True)
    
    print("\n")
    print(summary_df.to_string(
        index=False,
        col_space=8,
        justify="center"
    ))

    return summary_df
print("‚úÖ Fungsi visualisasi dan analisis siap")
```

```python
# ============================
# MENJALANKAN SIMULASI
# ============================

print("DISCRETE EVENT SIMULATION - SISTEM PIKET OMPRENG IT DEL")
print("="*60)

# ============================
# SKENARIO UTAMA
# ============================
results, df, model = run_scenario(
    num_meja=60,        # 60 meja
    petugas_lauk=3,     # isi lauk
    petugas_angkut=2,   # angkut ompreng
    petugas_nasi=2      # isi nasi
)

# ============================
# VISUALISASI HASIL
# ============================
visualize_results(df, model, " (Skenario Utama)")
```
